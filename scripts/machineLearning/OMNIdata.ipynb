```python
def retrieve_OMNI_data(dataset_name, start, end, vars=None):
    if vars is None:
        vars = vars = ['IMF','PLS','IMF_PTS','PLS_PTS','percent_interp','Timeshift','RMS_Timeshift','Time_btwn_obs','F','BX_GSE','BY_GSE','BZ_GSE','BY_GSM','BZ_GSM','RMS_SD_B','RMS_SD_fld_vec','flow_speed','Vx','Vy','Vz','proton_density','T','Pressure','E','Beta','Mach_num','Mgs_mach_num','x','y','z','BSN_x','BSN_y','BSN_z','AE_INDEX','AL_INDEX','AU_INDEX','SYM_D','SYM_H','ASY_D','ASY_H','PC_N_INDEX','PR-FLX_10','PR-FLX_30','PR-FLX_60']

    cdas = CdasWs()
    status, data = cdas.get_data(dataset_name, vars, start, end)

    df = None

    if status['http']['status_code'] == 200:
        df = data.to_dataframe()

    return df
```

```python
import pandas as pd
import numpy as np
```

```python
omni_data = pd.read_csv("raw/OMNI_2010_2023.csv")
```

```python
def fill_default_with_nan(df):
    dict_mag = {9999: pd.NA}
    dict_vel = {99999: pd.NA}
    dict_den = {999: pd.NA}
    dict_pre = {99: pd.NA}
    dict_ind = {99999: pd.NA}

    columns = ['F', 'BX_GSE', 'BY_GSM', 'BZ_GSM', 'flow_speed', 'proton_density'
               'Pressure', 'SYM_H', 'AE_INDEX', 'AL_INDEX', 'AU_INDEX']

    columns_to_default = {
        'F': dict_mag,
        'BX_GSE': dict_mag,
        'BY_GSM': dict_mag,
        'BZ_GSM': dict_mag,
        'flow_speed': dict_vel,
        'proton_density': dict_den,
        'Pressure': dict_pre,
        'SYM_H': dict_ind,
        'AE_INDEX': dict_ind,
        'AL_INDEX': dict_ind,
        'AU_INDEX': dict_ind
    }

    for col in columns_to_default:
        default, replace = list(columns_to_default[col].items())[0]
        df[col][df[col] == default] = replace

    return df
```

```python
omni_data_filled_default = fill_default_with_nan(omni_data)
# data['AL_INDEX'][data['AL_INDEX'] == 99999] = pd.NA
omni_data_filled_default
```

```python
def clean_up(df):
    df.rename(columns = {'Epoch':'DateTime'}, inplace = True)
    nan_cols = df.isnull().any()
    nan_cols = list(nan_cols[nan_cols == True].keys())

    df[nan_cols] = df[nan_cols].interpolate(method ='linear', axis=0, limit_direction='both')

    if df.isnull().any().sum() == 0:
        file_path = BASE_PATH + "clean/OMNI_2010_2023.csv"
        df.to_csv(BASE_PATH + "clean/OMNI_2010_2023.csv")

        print("The cleaned data is saved at:", file_path)

        return df

    return None
```

```python
cleaned_omni_data = clean_up(omni_data_filled_default)
```

```python
def process_and_resample_clockwise_data(file_name):
    raw_data = pd.read_csv(BASE_PATH + f'raw/{file_name}')
    raw_data.rename(columns = {'Date_UTC':'DateTime'}, inplace = True)

    raw_data['CLOCK_ANGLE_GSM'][raw_data['CLOCK_ANGLE_GSM'] == 999999.00] = pd.NA
    raw_data = raw_data.interpolate(method ='linear', axis=0, limit_direction='both')
    raw_data['DateTime'] = pd.to_datetime(raw_data['DateTime'])
    raw_data.set_index('DateTime', inplace=True)

    resample_data = raw_data.resample('5T').mean().interpolate()

    print(len(resample_data))

    assert(resample_data.isna().any().sum() == 0)
    print("Number of NaN:", resample_data.isna().any().sum())

    resample_data.to_csv(BASE_PATH + f'clean/{file_name}')

    return resample_data
```

```python
for i in range(2010, 2024):
    process_and_resample_clockwise_data(f'IMF_CLOCK_ANGLE_{i}.csv')
```
