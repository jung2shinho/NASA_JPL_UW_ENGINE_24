```python
import pandas as pd
import numpy as np
import pywt
import matplotlib.pyplot as plt

#use your own file path, here is an example
df = pd.read_csv('C:\\Users\\19606\\Desktop\\GMDandGIC\\GICdata\\2023E03_10212.csv')

# Extract GIC signal
gic_signal = df['GICMeasured'].values 

# wavelet analysis
level = 4  # The number of layers of wavelet decomposition
wavelet = 'haar'  # Select the Haar wavelet basis function

coeffs = pywt.wavedec(gic_signal, wavelet, level=level)

# Plot the original signal and wavelet coefficients
plt.figure(figsize=(12, 6))
plt.subplot(level + 2, 1, 1)
plt.plot(gic_signal)
plt.title('Original GIC Signal')

# Plot time-frequency diagram of wavelet coefficients 
for i in range(1, level + 1):
    plt.subplot(level + 2, 1, i + 1)
    plt.plot(coeffs[i - 1])
    plt.title(f'Wavelet Coefficients (Level {i})')

# reconstruct the signal
reconstructed_signal = pywt.waverec(coeffs, wavelet)

plt.subplot(level + 2, 1, level + 2)
plt.plot(reconstructed_signal)
plt.title('Reconstructed GIC Signal')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import pywt
import matplotlib.pyplot as plt

#use your own file path, here is an example
df = pd.read_csv('C:\\Users\\19606\\Desktop\\GMDandGIC\\GICdata\\2023E03_10212.csv')

# Extract GIC signal
gic_signal = df['GICMeasured'].values 

# wavelet analysis
level = 5  # The number of layers of wavelet decomposition
wavelet = 'haar'  # Select the Haar wavelet basis function

coeffs = pywt.wavedec(gic_signal, wavelet, level=level)

plt.figure(figsize=(20,15))

# Plot the original signal and wavelet coefficients
plt.subplot(level + 3, 1, 1)
plt.plot(gic_signal)
plt.title('Original GIC Signal')

# Plot time-frequency diagram of wavelet coefficients 
for i in range(level):
    plt.subplot(level + 3, 1, i + 2)
    plt.specgram(coeffs[i], cmap='viridis', Fs=1, NFFT=100, noverlap=50)
    plt.title(f'Wavelet Coefficients (Level {i+1})')
    plt.colorbar()

# reconstruct the signal
reconstructed_signal = pywt.waverec(coeffs, wavelet)


plt.subplot(level + 3, 1, level + 3)
plt.plot(reconstructed_signal)
plt.title('Reconstructed GIC Signal')

plt.subplots_adjust(hspace=1)
# plt.tight_layout()
plt.show()

```
